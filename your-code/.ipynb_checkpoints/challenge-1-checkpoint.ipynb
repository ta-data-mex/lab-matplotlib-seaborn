#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
# en una misma gráfica: 
plt.plot(x, y, label='(x,y)')
plt.plot(x, z, label='(x,z)')

plt.xlabel('x')
plt.ylabel('y,z')

plt.title("figure")
plt.legend()
plt.show()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
# using subplots

figura, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')  

ax2.plot(x,z)
ax2.set_title('Convex Plot')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')

plt.show()
```

```python
# also, there is a third way (which is basically the second way with different style)
figura, [ax1, ax2] = plt.subplots(2,1)
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')  

ax2.plot(x,z)
ax2.set_title('Convex Plot')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')

plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
# figsize=(15,15)

figura, [ax1, ax2] = plt.subplots(1,2,figsize=(22,12))
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')  

ax2.plot(x,z)
ax2.set_title('convex plot')
ax2.set_xlabel('x label')
ax2.set_ylabel('z label')

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
# normalizar las variables usando escala logarítmica: 

plt.xscale('log')
plt.yscale('log')

plt.plot(x, x**2, label='parábola')
plt.plot(x, np.e**x, label='exponencial')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("figure 2")
plt.legend(('$y = x ^ 2$','$y = e^x$'))
plt.show()
```
