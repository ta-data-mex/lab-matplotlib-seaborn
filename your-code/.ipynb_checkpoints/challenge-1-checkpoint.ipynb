#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

#Variables a graficar
x = np.arange(0,100)
y = x*2
z = x**2

plt.style.use('bmh')
plt.figure(figsize=(10 ,5))

plt.figure(facecolor='lightgrey')
plt.subplot(1,2,1)
plt.plot(x, y)
plt.subplot(1,2,2)
plt.plot(x, z)


```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)

plt.style.use('grayscale')
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

ax[0].title.set_text('Gráfico 1')
ax[0].plot(x, y) #para la linea

ax[1].title.set_text('Gráfico 2')
ax[1].scatter(x, z)  #para los puntos

plt.show();

```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here

plt.style.use('grayscale')
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 6), sharex=True)

ax[0].set_xlabel('eje x')
ax[0].set_ylabel('eje y')
ax[0].plot(x, y) #para la linea

ax[1].set_xlabel('eje x')
ax[1].set_ylabel('eje y')
ax[1].plot(x, z)  #para los puntos
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)


plt.style.use('grayscale')
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 6), sharex=True)

ax[0].set_xscale('linear')
ax[0].set_xlabel('eje x')
ax[0].set_ylabel('eje y')
ax[0].plot(x, y) #para la linea

ax[1].set_yscale('log')
ax[1].set_xlabel('eje x')
ax[1].set_ylabel('eje y')
ax[1].plot(x, z)  #para los puntos
plt.show();
```
