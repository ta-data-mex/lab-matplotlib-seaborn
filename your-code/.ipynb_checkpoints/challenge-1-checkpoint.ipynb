#### Import all the libraries that are necessary.

```python
# Import libraries here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Your code here-1st way (call `subplots` twice using the `index` parameter)

fig, axs = plt.subplots(2)

axs[0].plot(x, y)
axs[0].set_title('X vs. Y – Linear Relationship')

axs[1].plot(x, z)
axs[1].set_title('X vs. Z – Square Relationship')

fig.tight_layout(pad=1.5)
```

```python
fig, axs = plt.subplots(2)

plt.subplot(211, title='X vs. Y – Linear Relationship').plot(x, y)
plt.subplot(212, title='X vs. Z – Square Relationship').plot(x, z)

fig.tight_layout(pad=1.5)
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# Your code here

fig, axs = plt.subplots(2, figsize=(8, 6))

plt.subplot(211, title='X vs. Y – Linear Relationship').plot(x, y)
plt.subplot(212, title='X vs. Z – Square Relationship').plot(x, z)

fig.tight_layout(pad=1.5)
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# Your code here

fig, axs = plt.subplots(2, figsize=(8, 6))

plt.subplot(211, title='X vs. Y – Linear Relationship', xlabel='x-label', ylabel='y-label').plot(x, y)
plt.subplot(212, title='X vs. Z – Square Relationship', xlabel='x-label', ylabel='z-label').plot(x, z)

fig.tight_layout(pad=1.5)
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# Your code here

x = np.arange(0,100)
y = x**2
z = np.exp(x)
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('x-label')
ax1.set_ylabel('exp', color=color)
ax1.plot(x, y, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(x, z, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # Otherwise the right y-label is slightly clipped

plt.show()
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

```python
# Your code here

vehicles = pd.read_csv(r'/Users/diegoperezo97/Documents/Ironhack – Data Analytics Bootcamp/Module 2/Week 4/Day 3/lab-matplotlib-seaborn/your-code/vehicles.csv')
vehicles.head(5)
```

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
plt.scatter(vehicles['Combined MPG'], vehicles['Highway MPG'])
plt.title("'Combined MPG' vs. 'Highway MPG' ")
plt.xlabel('Combined MPG')
plt.ylabel('Highway MPG')
plt.show()
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# Your code here

plt.boxplot(vehicles['CO2 Emission Grams/Mile'])
plt.title("CO2 Emission Grams/Mile")
plt.xlabel('CO2 Emission Grams/Mile')
plt.ylabel('Count')
plt.show()
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# Your code here

plt.hist(vehicles['Fuel Barrels/Year'], bins=20)
plt.title("Frecuency of Fuel Barrels/Year")
plt.xlabel('Fuel Barrels/Year')
plt.ylabel('Frecuency')
plt.show()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
vehicles_pivot = pd.pivot_table(
    vehicles,
    values=['City MPG'],
    index=['Fuel Type'],
    aggfunc={'City MPG': 'mean'}
).reset_index()

vehicles_pivot
```

```python
plt.bar(vehicles_pivot['Fuel Type'], height=vehicles_pivot['City MPG'])
plt.xticks(rotation=90)
plt.title("Average City MPG vs Fuel Type")
plt.xlabel('Fuel Type')
plt.ylabel('Average City MPG')
plt.show()
```
