#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

```python
# your code here
# your code here
# your code here-1st way (call `subplots` twice using the `index` parameter)
#Lineplot

#x = top_10_botellas['item_description']
#y = top_10_botellas['sale_usd']

plt.style.use('ggplot')
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

ax[0].title.set_text('x vs y')
#ax[0].set_xticklabels([])
ax[0].plot(x, x**2)
plt.setp(ax[0].xaxis.get_majorticklabels(), rotation=90)
#x = 10.0**np.linspace(0.0, 2.0, 20)
#y = x**2.0

#y_botellas = top_10_botellas['bottles_sold']

ax[1].title.set_text('x vs z')
ax[1].scatter(x, z)
plt.xticks(rotation=45)
ax[0].set_xscale("log", nonpositive='clip')
ax[1].set_yscale("log", nonpositive='clip')


plt.show()
```
