#### Use the appropriate visulalization to show the distribution of the column `Age`

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
plt.style.use('ggplot')
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(10, 5))
ax[0].title.set_text('Histogram')
ax[0].hist(x_age, bins=10, density = True, 
         histtype ='bar',
         color = 'blue')
ax[1].hist(x_age, bins=20, density = True, 
         histtype ='bar',
         color = 'red')
ax[2].hist(x_age, bins=50, density = True, 
         histtype ='bar',
         color = 'gray')

plt.show();
```

#### Now in addition to the summary statistics, we also want to visualize the distribution of `Age`. Choose the appropriate plot to sow below.

```python
#your code here
fig, (ax_box, ax_hist) = plt.subplots(2, sharex=True, figsize=(10,10),
                                      gridspec_kw={'height_ratios': (0.15, 0.85)})

sns.boxplot(data=titanic['Age'], ax=ax_box, orient='h')
sns.histplot(data=titanic['Age'], ax=ax_hist)

ax_box.set(xlabel='')
plt.show();
```
