#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplot` twice using the `index` parameter)
plt.figure()

plt.subplot(1,2,1)
plt.plot(x,y, c='r', linewidth=5,linestyle='--', marker='1', label='Linear:$2x$')
plt.title('Linear Function',fontsize=15)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis',fontsize=15)
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,z, c='g', linewidth=2,linestyle='-', marker='^', label='Quadratic:$x^2$')
plt.title('Squared Functions',fontsize=15)
plt.xlabel('x-axis', fontsize=15)
plt.ylabel('y-axis',fontsize=15)
plt.legend()

plt.subplots_adjust(wspace=1, right=1)

plt.show()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y, c='r', linewidth=5,linestyle='--', marker='1', label='Linear:$2x$')
ax1.set_title('Linear Function')
ax1.legend()
ax1.set_xlabel('x-axis', fontsize=15)
ax1.set_ylabel('y-axis', fontsize=15)
ax2.plot(x,z, c='g', linewidth=2,linestyle='-', marker='^', label='Quadratic:$x^2$')
ax2.set_title('Squared Function')
ax1.legend()
ax2.set_xlabel('x-axis', fontsize=15)
ax2.set_ylabel('y-axis', fontsize=15)


plt.subplots_adjust( right=1.2,wspace=1)
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x = np.linspace(0,5,1000) 
y1 = x**2
y2 = np.exp(x)

plt.figure()

plt.subplot(121)
plt.plot(x,y1, c ='r', linewidth = 1)
plt.plot(x,y2, c ='g', linewidth = 1)
plt.title('Squared and Exponential Functions')
plt.ylabel('y')
plt.xlabel('x')
plt.xscale('linear')
plt.yscale('linear')


plt.subplot(122)
plt.plot(x,y1, c ='r', linewidth = 1)
plt.plot(x,y2, c ='g', linewidth = 1)
plt.title('Squared and Exponential Functions')
plt.ylabel('y')
plt.xlabel('x')
plt.xscale('log')
plt.yscale('log')
```
